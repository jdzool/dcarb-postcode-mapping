```python
%matplotlib inline
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

path = './data/postcodes/GB_Postcodes/PostalSector.shp'
postal_sector = gpd.read_file(path)
```

```python
postal_sector.head()
```

```python
f, ax = plt.subplots(1, figsize=(20, 20))
ax = postal_sector.plot(ax=ax)
ax.set_title('Map of postcode boundaries in UK')
ax.set_axis_off()
plt.show()
```
